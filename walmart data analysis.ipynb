```python

df['Order Date'] = pd.to_datetime(df['Order Date'] , format = '%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'] , format = '%d-%m-%Y')
```

```python
df.dtypes
```

```python
print(df['Order Date'].min())
print(df['Order Date'].max())
```

```python
df.head()
```

```python
df['O_day'] = df['Order Date'].dt.day
df['O_Month'] = df['Order Date'].dt.month
df['O_Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
def quarter(month):
  if month in [4,5,6]:
    return 'Q1'
  elif month in [7,8,9]:
    return 'Q2'
  elif month in [10,11,12]:
    return 'Q3'
  else:
    return 'Q4'
```

```python
quarter(7)
```

```python
df['Quarter'] = df['O_Month'].apply(quarter)   #apply() is a inbuilt function used to apply an user defined function on a data column
```

```python
df.head()
```

calculate the duration company took to ship a product

```python
df['Ship Date'] - df['Order Date']
```

```python
df['Processing Time'] = (df['Ship Date'] - df['Order Date']).apply(lambda x : x.days)
```

```python
new_order = ['Order ID', 'Order Date', 'Ship Date', 'Processing Time' , 'O_day', 'O_Month', 'O_Year',
             'Quarter', 'Customer Name', 'Country', 'City', 'State', 'Category', 'Product Name', 'Sales',
             'Quantity','Profit']

df = df.reindex(columns = new_order)      #reindex() can be used to change data order
df.head()
```

Pivot Table

```python
table = pd.pivot_table(data = df,
                       index = 'Category',                        #grouping factor
                       values = ['Order ID', 'Sales', 	'Quantity', 	'Profit'],
                       aggfunc = {'Order ID' : pd.Series.nunique,
                                  'Profit' : np.mean,
                                  'Quantity' : np.sum,
                                  'Sales' : np.mean})

table.columns = ['Total Orders' , 'Average Profit' , 'Total Quantity' , 'Average Sales']

table = table.sort_values('Total Quantity' , ascending = False)

table
```

```python
plt.figure(figsize = (9,6))

plt.bar(table.index , table['Total Orders'] , color = 'brown')

plt.xticks(rotation = 90)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Total Orders of each Category')

plt.show()
```

```python
plt.figure(figsize = (12,10))

plt.subplot(2,2,1)
plt.bar(table.index , table['Total Orders'] , color = 'brown')
plt.title("Plot 1 - Total Orders")
plt.xticks(rotation = 90)

plt.subplot(2,2,2)
plt.bar(table.index , table['Average Profit'] , color = 'brown')
plt.title("Plot 2 - Average Profit")
plt.xticks(rotation = 90)

plt.subplot(2,2,3)
plt.bar(table.index , table['Total Quantity'] , color = 'brown')
plt.title("Plot 3 - Total Quantity")
plt.xticks(rotation = 90)

plt.subplot(2,2,4)
plt.bar(table.index , table['Average Sales'] , color = 'brown')
plt.title("Plot 4 - Average Sales")
plt.xticks(rotation = 90)


plt.tight_layout()
plt.show()
```

```python
table.plot.barh(figsize = (12,8))
```
